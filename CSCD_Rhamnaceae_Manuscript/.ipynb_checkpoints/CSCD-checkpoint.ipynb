# Calculate chemical structural and compositional dissimilarity

**Author**: Madeleine Ernst (mernst@ucsd.edu) and Ricardo Silva (rsilva@ucsd.edu) <br>
**Edited by**: - <br>
**Use case**: Calculate chemical structural and compositional dissimilarity (CSCD) proposed by Sedio and collaborators (2017) (https://esajournals.onlinelibrary.wiley.com/doi/full/10.1002/ecy.1689) <br>
<ul>
<li>**Feature table** (.csv) is a comma separated file with features in rows and samples in columns. This corresponds to the MZmine .csv output table, filtered for features with associated MS2 data. </li>
<li>**Edges table** (.tsv) is a tab separated file containing edge information retrieved from mass spectral molecular networking (Download Cytoscape Data, networkedges_selfloop)
</li>
<li>**Metadata table** (.txt) is a tab separated file with samples in rows and metadata groups in columns. The column containing the sample names <font color='red'>must</font> be named <font color='red'>'filename'</font>. For detailed instructions check https://bix-lab.ucsd.edu/display/Public/Metadata+table+in+GNPS#MetadatatableinGNPS-MetadatatableinGNPS </li>
</ul> <br>
**Outputs**: CSCD distance matrix, dendrogram and interactive PCoA plot. <br> <br>
**Dependencies**: R version 3.4.2 (2017-09-28) and libraries ggplot2_2.2.1, RColorBrewer_1.1-2, igraph_1.2.1, dendextend_1.6.0, randomForest_4.6-12, scatterplot3d_0.3-40, pcaMethods_1.70.0, Biobase_2.38.0, BiocGenerics_0.24.0

load feature table

```r
features <- read.csv("FeatureTable_Rhamnaceae.csv", check.names=FALSE, stringsAsFactors=F, row.names = 1)
head(features[,1:8]) #ChemicalClass_DistanceMetric/
```

load edges file

```r
cos <- read.delim('0b5e59c0ea534d31b7af98e084462c98.pairsinfo') #Mass2Motifs_2_MolecularNetwork/
head(cos) 
```

```r
tab <- t(apply(t(features), 1, function(x) x/sum(as.numeric(x))))
tStart <- Sys.time()
mcscd <- cscd(tab, cos, norm=FALSE)
tEnd <- Sys.time()
tEnd-tStart
```

```r
write.csv(mcscd,"mcscd.csv",quote=F)
```

```r
identical(rownames(mcscd),as.character(metadata$filename)[1:71])
```

```r
dend <-  mcscd %>% as.dist %>%
  hclust %>% as.dendrogram 
```

```r
genera <- gsub(' [A-z ]*', '' , rownames(mcscd)[order.dendrogram(dend)])
genera <- gsub('[[:digit:]]+', '', genera)
genera <- gsub("[^[:alnum:]]", " ", genera)
genera <- gsub("[[:blank:]]", "", genera)

SubgCols <- c("purple3", "orangered2","purple","slateblue1","black","black","steelblue",
             "tomato","red","royalblue4","rosybrown2","yellow","tan1","darkorange","green","blue")

true_species_cols <- SubgCols[as.numeric(as.factor(genera))]
```

```r
genera
```

```r
true_species_cols
```

```r
dend <-  mcscd %>% as.dist %>%
  hclust %>% as.dendrogram %>%
  set("branches_lwd", 1) %>%
  set("labels_colors","black") %>% 
  set("labels_cex", c(0.5)) %>% 
  color_branches(k=71,col=true_species_cols) 
plot(dend)
```

```r
ggd1 <- as.ggdend(dend)
ggplot(ggd1, horiz = TRUE,offset_labels=-.01)
```

save plot to file

```r
pdf("CSCD_dendrogram_genera.pdf",width=6,height=7, bg="transparent") 
ggd1 <- as.ggdend(dend)
ggplot(ggd1, horiz = TRUE,offset_labels=-.01)
dev.off()
```

```r
genera2 <- gsub(' .+$', '' , metadata$filename)
cols <- SubgCols[as.numeric(as.factor(genera2))]
```

```r
d <- mcscd
d
```

```r
dict <- cbind(rownames(mcscd), genera, cols)
dict
```

```r
d <- cbind(rownames(d), d)
write.table(d, "dist_cscd.txt", sep="\t", row.names = FALSE, quote = FALSE)
```

```r
mapping <- data.frame(SampleID=dict[,1], BarcodeSequence="GATACA",   LinkerPrimerSequence="GATACA",  dict[,2], dict[,3], Description="Metabolome")
colnames(mapping) <- c("SampleID", "BarcodeSequence", "LinkerPrimerSequence",  "Genera", "Cols","Description")

write.table(matrix(c(paste0("#", colnames(mapping)[1]), colnames(mapping)[-1]), nrow=1), "mapping_file.txt", row.names = FALSE, col.names = FALSE, sep="\t", quote = FALSE)
write.table(mapping, "mapping_file.txt", row.names = FALSE, col.names = FALSE, sep="\t", append = TRUE, quote = FALSE)
head(mapping)
```
